```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import matplotlib
```

```python
income=[1.47,1.62,1.78,1.94,2.38,2.60]
year=range(2012,2018)
year
```

```python
data=pd.DataFrame({'Income':income},index=year)
```

```python
data
```

```python
data.plot(title='Income chart',color='k',linestyle=':',marker='s',ylim=(0,3),grid=True)
plt.annotate('Highest!',xy=(2017,2.60),xytext=(2015,2.55),arrowprops=dict(arrowstyle='->'))
plt.ylabel('Income(RMB Ten Thousand)',fontsize=12)
plt.xlabel('Year',fontsize=12)
plt.savefig('Income chart')
plt.show()
```

```python
fig=plt.figure(figsize=(6,5)) #figsize定义图形大小
ax1=fig.add_subplot(2,2,1)   #创建子图1 
ax1.plot(data)#用AxesSubplot绘制折线图
plt.title('line chart')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income')
ax2=fig.add_subplot(2,2,2)   #创建子图2 
data.plot(kind='box',fontsize='small',xticks=[],ax=ax2) #用pandas绘柱状图
plt.title('box-whisker plot')
plt.xlabel('2012-2017')
plt.ylabel('Income')
ax3=fig.add_subplot(2,1,2)   #创建子图3 
data.plot(kind='bar',use_index=True,fontsize='small',ax=ax3)#用pandas绘柱状图
plt.title('bar chart')
plt.xlabel('Year')
plt.ylabel('Income')
plt.subplots_adjust(hspace = 0.4,wspace=0.4,right = 0.75,bottom=0)#天知道我为了复刻课本上的图调了这玩意多久
#坑爹啊，index如果是字符串类型的话就做不出来了
plt.show()
```
