```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cdata = pd.read_csv('data/bankpep.csv')      #读文件
cdata.head()
```

```python
cdata['age'].plot(kind='hist',bins=6,normed=True,title='Customer Age')
cdata['age'].plot(kind='kde',title="Customer Age",style='k--',ylim=[0,0.025])
```

```python
plt.show()
```

```python
cdata.plot(kind='scatter',x='age',y='income',title='Customer Income',grid=True,xlim=[0,80],marker='o',label='(Age,Income)',color='m')
```

```python
plt.show()
```

```python
data=cdata[['age','income','children']]
pd.plotting.scatter_matrix(data,diagonal='hist',color='m')
plt.show()
```

```python
data=cdata.groupby(['region'])
#mean=data.aggregate({'income':np.mean})
mean=data['income'].mean()
std=data.aggregate({'income':np.std})
```

```python
mean.plot(kind='bar',yerr=std,color='cadetblue',title = 'Customer Income',rot=45)
```

```python
plt.show()
```

```python
fig=plt.figure(figsize=(6,6))
ax1=fig.add_subplot(2,2,1)
sdata=cdata['sex'].value_counts()
sdata.plot( kind='pie', title='Customer Sex',fontsize=14,shadow=True,startangle=60, autopct='%1.1f%%')
```

```python
ax2=fig.add_subplot(2,2,2)
scdata=cdata[cdata['car']=='YES']['sex'].value_counts()
scdata.plot(kind='pie',title='Customer Car Sex',fontsize=14,shadow=True,startangle=60, autopct='%1.1f%%',counterclock = False)
```

```python
ax3=fig.add_subplot(2,2,3)
ccdata=cdata['children'].value_counts()
ccdata.plot(kind='pie',title='Customer Children',fontsize=14,shadow=True,startangle=60, autopct='%1.1f%%')
```

```python
plt.show()
```

```python
cdata1 = cdata[['sex','income']]
cdata1.boxplot(by='sex',figsize=(6,6))
plt.show()
```
