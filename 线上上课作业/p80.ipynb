# 高铁

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data/high-speed rail.csv',index_col='Country')
```

```python
print(data)
```

```python
data1=data['Operation']
```

```python
data1.plot(kind='bar',title = 'Operation Mileage', rot=45)
plt.annotate('longest!',xy=(0,20000),xytext=(1,20000),arrowprops=dict(arrowstyle='->'))
plt.ylabel('Mileage(km)',fontsize=12)
plt.show()
```

```python
data2=data[['Operation','Under-construction','Planning']]
data2.plot(kind='barh',stacked=True,title = 'Global trends of high-speed rail')
plt.xlabel('Mileage(km)',fontsize=12)
plt.show()
```

```python
data1.name=''
data1.plot( kind='pie', figsize=(6,6), title='Operation Mileage',fontsize=14,explode=[0.1,0,0,0,0,0],shadow=False,startangle=60, autopct='%1.1f%%')
plt.show()
```
